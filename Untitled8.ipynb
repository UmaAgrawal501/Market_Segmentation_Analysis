```python
def perform_kMeans(data, k_values, n_init=10):
    results = {}
    for k in k_values:
        kMeans = KMeans(n_clusters=k, n_init=n_init, random_state=1234)
        labels = kMeans.fit_predict(data)
        score = silhouette_score(data, labels)
        results[k] = {
            'kMeans': kMeans,
            'labels': labels,
            'score': score
        }
    return results
```

```python
def calculate_silhouette_scores(data, k_values, n_init=10):
    scores = []
    for k in k_values:
        kmeans = KMeans(n_clusters=k, n_init=n_init, random_state=1234)
        labels = kmeans.fit_predict(data)
        score = silhouette_score(data, labels)
        scores.append(score)
    return scores
```

```python
def bootstrapped_clustering(data, k_range, nrep, nboot):
    results = []
    for k in k_range:
        kmeans_scores = []
        for _ in range(nboot):
            # Bootstrap sample
            bootstrap_sample = resample(data, replace=True, n_samples=data.shape[0])
            # Fit KMeans clustering
            kmeans = KMeans(n_clusters=k, random_state=1234).fit(bootstrap_sample)
            # Compute silhouette score
            score = silhouette_score(bootstrap_sample, kmeans.labels_)
            kmeans_scores.append(score)
        # Average silhouette score for current k
        results.append((k, np.mean(kmeans_scores)))
    return results
```

```python
MD_x = np.random.binomial(1, 0.5, (100, 5))
```

```python
def compute_aic(n_samples, log_likelihood, n_params):
    return 2 * n_params - 2 * log_likelihood

def compute_bic(n_samples, log_likelihood, n_params):
    return np.log(n_samples) * n_params - 2 * log_likelihood
```

```python
results = []
n_clusters_range = range(2, 9)
```

```python
for k in n_clusters_range:
    gmm = GaussianMixture(n_components=k, covariance_type='diag', random_state=1234)
    gmm.fit(MD_x)
    log_likelihood = gmm.score(MD_x) * MD_x.shape[0]
    n_params = gmm.n_components * (2 * MD_x.shape[1] - 1)
    aic = compute_aic(MD_x.shape[0], log_likelihood, n_params)
    bic = compute_bic(MD_x.shape[0], log_likelihood, n_params)
    labels = gmm.predict(MD_x)
    cluster_probas = gmm.predict_proba(MD_x)

    # Calculate ICL using a more appropriate method
    icl = bic + np.sum(cluster_probas * np.log(cluster_probas))

    results.append({
        'k': k,
        'logLik': log_likelihood,
        'AIC': aic,
        'BIC': bic,
        'ICL': icl
    })
results_df = pd.DataFrame(results)

print(results_df)
```
